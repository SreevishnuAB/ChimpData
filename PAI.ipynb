```python
import pandas as pd
from typing import List, Dict
```

```python
df = pd.read_csv("drive/MyDrive/Colab Notebooks/chimp_data_with_nn_standardised_final_istg.csv", dtype="str")
```

```python
df.columns
```

```python
df_pai = df.loc[:, ['Month', 'fuata_follow', 'neighbours_normal:nearest_neighbour', 'neighbours_normal:other_neighbours']]
df_pai = df_pai.fillna('')
df_pai = df_pai.rename(columns={"fuata_follow": "A", "neighbours_normal:nearest_neighbour": "B", "neighbours_normal:other_neighbours": "party"})
df_pai
```

```python
df_pai['A'] = df_pai['A'].str.lower()
df_pai['B'] = df_pai['B'].str.lower()
df_pai['party'] = df_pai['party'].str.lower()
df_pai
```

```python
valid_names: List = list(set(df_pai['A'].unique()))
valid_names.remove("unknown")
valid_names
```

```python
to_be_cleaned = list(set(df_pai['B'].unique()) - set(df_pai['A'].unique()))
to_be_cleaned.sort(key=len)
to_be_cleaned
```

```python
valid_names.extend(to_be_cleaned[0:6])
valid_names
```

```python
df_pai.loc[df_pai['B'] == "bingwabingwa", 'B'] = "bingwa"
df_pai[df_pai['B'] == "bingwabingwa"] # sanity check
```

```python
df_clean = pd.DataFrame()

for name in valid_names:
  df_clean = pd.concat([df_clean, df_pai[df_pai['B'] == name]])

df_clean
```

```python
df_pai_wo_0: pd.DataFrame = df_clean.loc[(df_clean['A'] != '0') & (df_clean['B'] != '0'), :]
df_pai_wo_0 = df_pai_wo_0[~((df_pai_wo_0['A'] == 'sanaa') & (df_pai_wo_0['B'] == 'sanaa'))]
df_pai_wo_0 = df_pai_wo_0[~((df_pai_wo_0['A'] == 'kinanda') & (df_pai_wo_0['B'] == 'kinanda'))]
df_pai_wo_0 = df_pai_wo_0[~((df_pai_wo_0['A'] == 'kinanda') & (df_pai_wo_0['B'] == 'kinanda'))]
df_pai_wo_0 = df_pai_wo_0[~((df_pai_wo_0['A'] == 'unknown') | (df_pai_wo_0['B'] == 'unknown'))]

df_pai_wo_0
```

```python
df_pai_wo_0['party'] = df_pai_wo_0['A'] + ',' + df_pai_wo_0['B'] + ',' + df_pai_wo_0['party']
df_pai_wo_0
```

```python
df_pai_wo_0.to_csv("drive/MyDrive/Colab Notebooks/chimp_data_cleaned.csv", header=True, index=False)
```

```python
df_pai_wo_0['Month'] = pd.to_numeric(df_pai_wo_0['Month'])
df_pai_wet_season = df_pai_wo_0[(df_pai_wo_0['Month'] >= 11) | (df_pai_wo_0['Month'] <= 4)]
df_pai_dry_season = df_pai_wo_0[(df_pai_wo_0['Month'] >= 5) & (df_pai_wo_0['Month'] <= 10)]
df_pai_wet_season
```

```python
gpby_wet_season = df_pai_wet_season.groupby(['A', 'B'])
unique_pairs_wet_season = map(lambda item: frozenset(item), list(gpby_wet_season.groups.keys()))
list_unique_pairs_wet_season: List = [list(i) for i in set(unique_pairs_wet_season)]
list_unique_pairs_wet_season.sort(key=lambda x: x[0])
unique_pair_count_wet_season: Dict = {}

for a, b in list_unique_pairs_wet_season:
  if a not in unique_pair_count_wet_season:
    unique_pair_count_wet_season[a] = {}
  unique_pair_count_wet_season[a][b] = 0

unique_pair_count_wet_season
```

```python
gpby_dry_season = df_pai_dry_season.groupby(['A', 'B'])
unique_pairs_dry_season = map(lambda item: frozenset(item), list(gpby_dry_season.groups.keys()))
list_unique_pairs_dry_season: List = [list(i) for i in set(unique_pairs_dry_season)]
list_unique_pairs_dry_season.sort(key=lambda x: x[0])
unique_pair_count_dry_season: Dict = {}

for a, b in list_unique_pairs_dry_season:
  if a not in unique_pair_count_dry_season:
    unique_pair_count_dry_season[a] = {}
  unique_pair_count_dry_season[a][b] = 0

unique_pair_count_dry_season
```

```python
unique_individuals_wet_season = set()

for pairs in list_unique_pairs_wet_season:
  unique_individuals_wet_season.update(pairs)

list_unique_individuals_wet_season = list(unique_individuals_wet_season)

list_unique_individuals_wet_season
```

```python
unique_individuals_dry_season = set()

for pairs in list_unique_pairs_dry_season:
  unique_individuals_dry_season.update(pairs)

list_unique_individuals_dry_season = list(unique_individuals_dry_season)

list_unique_individuals_dry_season
```

```python
df_pai_wet_season.loc[:, 'party'] = df_pai_wet_season.loc[:, 'party'].str.rstrip(',')
df_pai_wet_season[df_pai_wet_season['party'] == '']  # sanity check
```

```python
df_pai_dry_season.loc[:, 'party'] = df_pai_dry_season.loc[:, 'party'].str.rstrip(',')
df_pai_dry_season[df_pai_dry_season['party'] == '']  # sanity check
```

```python
list_party_wet_season = list(df_pai_wet_season['party'].values)
list_party_wet_season = list(map(lambda party: party.split(","), list_party_wet_season))
list_party_wet_season
```

```python
list_party_dry_season = list(df_pai_dry_season['party'].values)
list_party_dry_season = list(map(lambda party: party.split(","), list_party_dry_season))
list_party_dry_season
```

```python
dict_party_freq_wet_season = {individual: {} for individual in unique_individuals_wet_season}
dict_party_freq_wet_season
```

```python
dict_party_freq_dry_season = {individual: {} for individual in unique_individuals_dry_season}
dict_party_freq_dry_season
```

```python
for individual in unique_individuals_wet_season:
   for party in list_party_wet_season:
    #  print(party)
     if individual in party:
       party_size = len(party)
       if party_size in dict_party_freq_wet_season[individual].keys():
         dict_party_freq_wet_season[individual][party_size] += 1
       else:
         dict_party_freq_wet_season[individual][party_size] = 1
dict_party_freq_wet_season
```

```python
for individual in unique_individuals_dry_season:
   for party in list_party_dry_season:
    #  print(party)
     if individual in party:
       party_size = len(party)
       if party_size in dict_party_freq_dry_season[individual].keys():
         dict_party_freq_dry_season[individual][party_size] += 1
       else:
         dict_party_freq_dry_season[individual][party_size] = 1
dict_party_freq_dry_season
```

```python
# for denominator terms
n_total_neighbours_individual_wet_season = {}

for individual, party_info in dict_party_freq_wet_season.items():
  t = 0
  for party_size, n_occurences in party_info.items():
    t += (n_occurences * (party_size - 1))
  n_total_neighbours_individual_wet_season[individual] = t

n_total_neighbours_individual_wet_season
```

```python
# for denominator terms
n_total_neighbours_individual_dry_season = {}

for individual, party_info in dict_party_freq_dry_season.items():
  t = 0
  for party_size, n_occurences in party_info.items():
    t += (n_occurences * (party_size - 1))
  n_total_neighbours_individual_dry_season[individual] = t

n_total_neighbours_individual_dry_season
```

```python
# numerator term 2

n_total_neighbours_all_wet_season = 0
for party in list_party_wet_season:
  party_size = len(party)
  n_total_neighbours_all_wet_season += (party_size * (party_size - 1))

n_total_neighbours_all_wet_season
```

```python
# numerator term 2

n_total_neighbours_all_dry_season = 0
for party in list_party_dry_season:
  party_size = len(party)
  n_total_neighbours_all_dry_season += (party_size * (party_size - 1))

n_total_neighbours_all_dry_season
```

```python
list(map(lambda x: list(x), df_pai_wo_0[['A', 'B']].values))
```

```python
pair_occurrence_wet_season: Dict = {}

for a, b in list_unique_pairs_wet_season:
  count = 0
  for party in list_party_wet_season:
    if a in party and b in party:
      count += 1
  if a not in pair_occurrence_wet_season.keys():
    pair_occurrence_wet_season[a] = {}
  pair_occurrence_wet_season[a][b] = count

pair_occurrence_wet_season

# list_unique_pairs
```

```python
pair_occurrence_dry_season: Dict = {}

for a, b in list_unique_pairs_dry_season:
  count = 0
  for party in list_party_dry_season:
    if a in party and b in party:
      count += 1
  if a not in pair_occurrence_dry_season.keys():
    pair_occurrence_dry_season[a] = {}
  pair_occurrence_dry_season[a][b] = count

pair_occurrence_dry_season

# list_unique_pairs
```

```python
pai_by_pairs_wet_season: List = []

for a, b in list_unique_pairs_wet_season:
  iab = pair_occurrence_wet_season[a][b]
  n_ai = n_total_neighbours_individual_wet_season[a]
  n_bi = n_total_neighbours_individual_wet_season[b]
  pai = (iab * n_total_neighbours_all_wet_season)/(n_ai * n_bi)
  pai_by_pairs_wet_season.append({'A': a, 'B': b, "PAI": pai})

pai_by_pairs_wet_season
```

```python
pai_by_pairs_dry_season: List = []

for a, b in list_unique_pairs_dry_season:
  iab = pair_occurrence_dry_season[a][b]
  n_ai = n_total_neighbours_individual_dry_season[a]
  n_bi = n_total_neighbours_individual_dry_season[b]
  pai = (iab * n_total_neighbours_all_dry_season)/(n_ai * n_bi)
  pai_by_pairs_dry_season.append({'A': a, 'B': b, "PAI": pai})

pai_by_pairs_dry_season
```

```python
df_pai_by_pairs_wet_season = pd.DataFrame(pai_by_pairs_wet_season)
df_pai_by_pairs_wet_season.to_csv("drive/MyDrive/Colab Notebooks/pai_by_pairs_wet_season.csv", header=True, index=False)
df_pai_by_pairs_wet_season
```
