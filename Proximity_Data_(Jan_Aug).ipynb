```python
import pandas as pd
df2 = pd.read_excel('drive/MyDrive/Colab Notebooks/2020 8-12.xlsx', dtype="str")
df2[df2['nearest_neighbour'] == "basekanot_visible"]
```

```python
df1 = pd.read_excel('drive/MyDrive/Colab Notebooks/2021 1-8.xlsx', dtype="str")
df2 = pd.read_excel('drive/MyDrive/Colab Notebooks/2020 8-12.xlsx', dtype="str")

df1['Year'] = "2021"
df2['Year'] = "2020"
df1 = df1.rename(columns={"Date": "Day"})
df1 = df1[["Year", "Month", "Day", "scan_time", "fuata_follow", "neighbours_normal:nearest_neighbour", "neighbours_normal:other_neighbours"]]
df2 = df2[["Year", "Month", "Day", "scan:time", "individual:id_focal", "nearest_neighbour", "neighbours:other_neighbours"]]

columns_df1 = list(df1.columns)
columns_df2 = list(df2.columns)

rename_dict = {k: v for k, v in zip(columns_df2, columns_df1)}

df2 = df2.rename(columns=rename_dict)
df2
```

```python
df2[df2['neighbours_normal:nearest_neighbour'] == "basekanot_visible"]
```

```python
df = pd.concat([df1, df2], ignore_index=True)
df[df["neighbours_normal:nearest_neighbour"] == "basekanot_visible"]
```

```python
df[df.index.duplicated()]
```

```python
df['Month'] = df['Month'].str.zfill(2)
df['Day'] = df['Day'].str.zfill(2)
df
```

```python
df2
```

```python
df_cleaned = df[~df['fuata_follow'].isnull()]
df_cleaned = df_cleaned[~(df['neighbours_normal:nearest_neighbour'].isnull() & df['neighbours_normal:other_neighbours'].isnull())]
df_cleaned
```

```python
df_cleaned[df_cleaned['neighbours_normal:other_neighbours'] == 'corona elisha imba maisha maua mosi wingu']
```

```python
df_cleaned['hour'] = ''
df_cleaned.loc[:, 'hour'] = df_cleaned['scan_time'].astype('str').str[:2]
df_cleaned['hour']
```

```python
# df_cleaned.loc[:, 'neighbours_normal:other_neighbours'] = ",".join(df_cleaned['neighbours_normal:other_neighbours'].str.split(' '))
# df_cleaned
df_cleaned.loc[~df_cleaned['neighbours_normal:other_neighbours'].isna(), 'neighbours_normal:other_neighbours'] = df_cleaned.loc[~df_cleaned['neighbours_normal:other_neighbours'].isna(), 'neighbours_normal:other_neighbours'].str.split(' ')
df_cleaned.loc[~df_cleaned['neighbours_normal:other_neighbours'].isna(), 'neighbours_normal:other_neighbours'] = df_cleaned.loc[~df_cleaned['neighbours_normal:other_neighbours'].isna(), 'neighbours_normal:other_neighbours'].agg(','.join)
df_cleaned
```

```python
df_cleaned['Date'] = df_cleaned['Year'] + '-' + df_cleaned['Month'] + '-' + df_cleaned['Day']
df_cleaned
```

```python
df_relevant_columns = df_cleaned[['Month', 'Date', 'hour', 'fuata_follow', 'neighbours_normal:nearest_neighbour', 'neighbours_normal:other_neighbours']]
gpby = df_cleaned.groupby(['Month', 'Date', 'hour', 'fuata_follow'])

df_grouped = pd.DataFrame()
for _, group in gpby:
  df_grouped = pd.concat([df_grouped, group])
  # print(group)

df_grouped.to_csv('chimp_data_filtered.csv', header=True)
```

```python
list_processed = []
for _, group in gpby:
  # print(_)
  set_nn = set(group.loc[~group['neighbours_normal:nearest_neighbour'].isna(), 'neighbours_normal:nearest_neighbour'].values)
  set_on = set(group.loc[~group['neighbours_normal:other_neighbours'].isna(), 'neighbours_normal:other_neighbours'].values)
  # print(set_on)
  first_row = group.iloc[0]
  first_row['neighbours_normal:nearest_neighbour'] = ','.join(str(v) for v in list(set_nn))
  first_row['neighbours_normal:other_neighbours'] = ','.join(str(v) for v in list(set_on))
  # print(first_row)
  list_processed.append(first_row)
df_processed = pd.DataFrame(list_processed)
df_processed
```

```python
df_to_group = df_processed[['Month', 'Date', 'hour', 'fuata_follow', 'neighbours_normal:nearest_neighbour', 'neighbours_normal:other_neighbours']]
gpby_fuata_hour = df_to_group.groupby(['fuata_follow', 'Month', 'Date', 'hour'])
df_formatted = pd.DataFrame()
for _, group in gpby_fuata_hour:
  df_formatted = pd.concat([df_formatted, group])
df_formatted = df_formatted[['Month', 'Date', 'hour', 'fuata_follow', 'neighbours_normal:nearest_neighbour', 'neighbours_normal:other_neighbours']]
df_formatted
```

```python
df_formatted.to_csv('chimp_data_fixed?.csv', header=True)
```

```python
gpby_with_nn = df_cleaned.groupby(['fuata_follow', 'neighbours_normal:nearest_neighbour', 'Date', 'hour'])

df_grouped_with_nn = pd.DataFrame()
for _, group in gpby_with_nn:
  df_grouped_with_nn = pd.concat([df_grouped_with_nn, group])

df_grouped_with_nn
```

```python
list_processed_nn = []
for _, group in gpby_with_nn:
  # print(_)
  set_on = set(group.loc[~group['neighbours_normal:other_neighbours'].isna(), 'neighbours_normal:other_neighbours'].values)
  set_nn = set(group.loc[~group['neighbours_normal:nearest_neighbour'].isna(), 'neighbours_normal:nearest_neighbour'].values)
  set_on = set_on - set_nn
  first_row = group.iloc[0]
  first_row['neighbours_normal:other_neighbours'] = ','.join(str(v) for v in list(set_on))
  # print(first_row)
  list_processed_nn.append(first_row)
df_processed_nn = pd.DataFrame(list_processed_nn)
df_processed_nn['scan_time'] = df_processed_nn['Date'] + " " + df_processed_nn['scan_time']
df_processed_nn = df_processed_nn[["Date", "scan_time",	"fuata_follow",	"neighbours_normal:nearest_neighbour",	"neighbours_normal:other_neighbours"]]
df_processed_nn
```

```python
df_processed_nn.to_csv('chimp_data_with_nn_standardised_final_istg.csv', header=True, index=False)
```
